```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paso de cada eje (0.3, 0.6, 0.9 ...)
nx, ny = 0.3, 0.3
x = np.arange(-5, 5, nx) # Valores de X desde -5 a 5, dividido en 0.3
y = np.arange(-5, 5, ny) # Valores de Y desde -5 a 5, dividido en 0.3

# rectangular grid with points
X, Y = np.meshgrid(x, y)
```

```python
# Derivadas (ec. dif.)
factor = 5 # Scale for Spirals
miu = [-1, 0 , 1] # Parametro de posicion (para puntos criticos)

saddlenode_bifurcation = (miu[1]-X**2, -Y, "$\dot{x} = \mu-x^2$ ; $\dot{y}=-y$") # Primer ecuacion
transcritical_bifurcation = (miu[0]*X-X**2, -Y, "$\dot{x} = \mu*x-x^2$ ; $\dot{y}=-y$") # Segunda ecuacion
pitchfork_bifurcation = (miu[2]*X-X**3, -Y, "$\dot{x} = \mu*x-x^3$ ; $\dot{y}=-y$") # Tercera ecuacion
hopf_bifurcation = ( miu[0]*X + factor * Y, -1*factor*X + miu[0]*Y, "$\dot{r} = r(\mu-r^2)$ ; $\dot{O}=-1$" ) # Hopf en forma rectangular

dx = saddlenode_bifurcation[0]
dy = saddlenode_bifurcation[1]
str_eq = saddlenode_bifurcation[2]
```

```python
# Normalizacion, para que se vean mejor las flechas
dyu = dy /np.sqrt(dx**2 + dy**2)
dxu = dx /np.sqrt(dx**2 + dy**2)
```

```python
# plot
plt.figure(figsize=(10,8))
plt.quiver(X, Y, dxu, dyu, color='purple')
plt.title('Direction Field for {}'.format(str_eq))
plt.show()
```
