```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
RESISTENCIA = 10e3 # 1kOhm
CAPACITANCIA = 1e-6 # 1uF
VSOURCE = 5 # 5 Volts
TAU = RESISTENCIA * CAPACITANCIA
V0 = 0
CHARGE_TIME = VSOURCE * TAU
```

```python
simul_time = np.arange(0, CHARGE_TIME * 2, 0.01 )
```

```python
voltages = []
for idx, t in enumerate(simul_time):
    factor = int(t/CHARGE_TIME)
    if t == 0 or t % CHARGE_TIME == 0:
        at = V0
    else:
        at = round(t - factor*CHARGE_TIME, 3) 
    voltages.append( VSOURCE * (1 - math.exp(-at/TAU)) )
```

```python
plt.plot(simul_time, voltages)
```
